```
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.goibibo.com/hotels/hotels-in-prayagraj-ct/"
headers = {
    'User-Agent': "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36"
    }
response = requests.request("GET", url, headers=headers)

data = BeautifulSoup(response.text, 'html.parser')
```

```
# find all the sections with specifiedd class name
cards_data = data.find_all('div', attrs={'class', 'HotelCardstyles__HotelCardInfoWrapperDiv-sc-1s80tyk-6 dfmysf'})

# total number of cards
print('Total Number of Cards Found : ', len(cards_data))

# source code of hotel cards
for card in cards_data:
    print(card)
```

```
print(cards_data[0])
```

```
scraped_data = []
for card in cards_data:
    hotel_details={}
    # get the room price
    hotel_price=card.find('p', attrs={'class': 'HotelCardstyles__CurrentPrice-sc-1s80tyk-27 czKsrL'})
    rating=card.find('span', attrs={'class': 'ReviewAndRatingsstyles__AverageReviewText-sc-1nxmeoo-8 MUGDq'})
    hotel_name=card.find('a', attrs={'class': 'HotelCardstyles__HotelNameSeoAnchor-sc-1s80tyk-12 kZJdUo'})
    address=card.find('div', attrs={'class': 'PersuasionHoverTextstyles__TextWrapperSpan-sc-1c06rw1-14 hbBHvg'})
    desc=card.find('div', attrs={'class': 'PersuasionHoverTextstyles__TextWrapperSpan-sc-1c06rw1-14 jLajgr'})
    discount=card.find('div', attrs={'class': 'OfferTagstyles__PercentageOffWrapperDiv-sc-16zl30v-10 fShEyB'})
    temp_link="http://www.goibibo.com"+hotel_name['href']
    final_link=""
    for ch in temp_link:
      if(ch=='?'):
        break
      final_link=final_link+ch
    address_desc="Not Available"
    if(type(address)!=type(None)):
      if(type(address.span)!=type(None)):
        address_desc=address.span.text

    desc_text="Not Available"
    if(type(desc)!=type(None)):
      if(type(desc.span)!=type(None)):
        desc_text=desc.span.text

    facility=card.find_all('p', attrs={'class', 'AmenitiesListstyles__TextWrapper-sc-19dqtu1-7 hoxxAR'})
    facility_list=[]
    for ele in facility:
      facility_list.append(ele.text)

    print(hotel_name['content'],final_link,address_desc,desc_text)
    print(discount.text)
    print(hotel_price.text)

    hotel_details['Name']=hotel_name['content']
    hotel_details['Link']=final_link
    hotel_details['Address']=address_desc
    hotel_details['Description']=desc_text
    hotel_details['Price per night']=hotel_price.text
    hotel_details['Discount']=discount.text

    scraped_data.append(hotel_details)
    
    #hotel_response=requests.request("GET",final_link, headers=headers)
    #hotel_data=BeautifulSoup(hotel_response.text, 'html.parser')

dataFrame = pd.DataFrame.from_dict(scraped_data)

# save the scraped data as CSV file
dataFrame.to_csv('hotels_data.csv', index=False)


#OfferTagstyles__PercentageOffWrapperDiv-sc-16zl30v-10 fShEyB
#AmenitiesListstyles__TextWrapper-sc-19dqtu1-7 hoxxAR
#PersuasionHoverTextstyles__TextWrapperSpan-sc-1c06rw1-14 jLajgr
#PersuasionHoverTextstyles__TextWrapperSpan-sc-1c06rw1-14 hbBHvg
#ReviewAndRatingsstyles__AverageReviewText-sc-1nxmeoo-8 MUGDq
#HotelCardstyles__CurrentPrice-sc-1s80tyk-27 czKsrL
#HotelCardstyles__HotelNameWrapperDiv-sc-1s80tyk-11 hiiHjq
```
